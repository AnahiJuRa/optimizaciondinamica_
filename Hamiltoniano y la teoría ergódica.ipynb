# Optimización Dinámica  

La optimización dinámica es una rama de la matemática aplicada que se ocupa de minimizar o maximizar una funcional sobre trayectorias (o arcos) que satisfacen ciertas restricciones dinámicas. Estas restricciones suelen formularse como ecuaciones diferenciales controladas o inclusiones diferenciales (Franke & Stroustrup, 2019). Este enfoque es ampliamente utilizado en diversas disciplinas, como la economía, la ingeniería y la biología, para modelar y resolver problemas en los que las decisiones deben tomarse a lo largo del tiempo.    

Dentro de la optimización dinámica, los problemas de control óptimo constituyen una subcategoría importante. Estos problemas emplean principios fundamentales como el Principio del Máximo de Pontryagin y la programación dinámica para encontrar soluciones óptimas (Sieniutycz & Jezowski, 2009). En economía, los métodos de optimización dinámica más comunes incluyen:    
- El cálculo de variaciones,    
- La teoría de control óptimo,    
- La programación dinámica en su forma discreta.    

Los avances recientes en algoritmos han mejorado significativamente el rendimiento en la resolución de problemas de optimización dinámica. Por ejemplo, un algoritmo mejorado para estrategias simultáneas, que combina métodos de puntos interiores con un novedoso método de búsqueda de línea basado en filtros, ha demostrado ser altamente eficiente (Biegler, Cervantes, & Wächter, 2002).    

En este contexto, la optimización dinámica no solo proporciona herramientas teóricas para analizar problemas complejos, sino que también ofrece soluciones prácticas para aplicaciones en tiempo real, como la gestión de recursos, el diseño de procesos químicos y la planificación económica.A continuación dejaré usando un ejemplo en Modelo de Crecimiento de Ramsey utilizando el Halmitoniano y la teoría ergódica.  



 



En el modelo de crecimiento de Ramsey abordamos un problema de optimización dinámica en el que buscamos maximizar una función de utilidad intertemporal sujeta a restricciones dinámicas. A continuación, se presenta la teoría paso a paso.  

---  

### El planteamiento del Problema  

Queremos maximizar la utilidad intertemporal de un agente representativo, dada por:  
$  
\max_{c(t)} \int_0^\infty e^{-\rho t} \ln(c(t)) \, dt,  
$  
donde:  
- $c(t)$ es el consumo en el tiempo $t$,  
- $\rho > 0$ es la tasa de descuento subjetiva, que mide la preferencia temporal del agente,  
- $\ln(c(t))$ es la función de utilidad instantánea, que refleja la satisfacción obtenida por el consumo en cada instante.  

### Restricciones del problema:  
1. **Ecuación de movimiento del capital**:  
   $  
   \dot{k}(t) = Ak(t) - c(t),  
   $  
   donde:  
   - $k(t)$ es el capital en el tiempo $t$,  
   - $A > 0$ es la productividad del capital,  
   - $c(t)$ es el consumo en el tiempo $t$,  
   - $\dot{k}(t)$ es la tasa de cambio del capital.  

2. **Condición de no divergencia del capital**:  
   $  
   \lim_{t \to \infty} k(t) e^{-\rho t} = 0.  
   $  
   Esta condición asegura que el capital no crezca indefinidamente y que el sistema sea económicamente viable.  

---  

### Ahora usando el  método del Hamiltoniano  

Para resolver este problema, utilizamos el **método del control óptimo**. Este método introduce un multiplicador de Lagrange dinámico $\lambda(t)$ para incorporar la restricción dinámica en el problema de maximización.  

### Definición del Hamiltoniano:  
El Hamiltoniano se define como:  
$  
H = e^{-\rho t} \ln(c(t)) + \lambda(t) \big[Ak(t) - c(t)\big],  
$  
donde:  
- $e^{-\rho t} \ln(c(t))$ es la utilidad instantánea descontada,  
- $\lambda(t)$ es el multiplicador de Lagrange asociado a la restricción dinámica.  

---  



Para encontrar la solución óptima, derivamos las **condiciones de primer orden** del problema. Estas condiciones son:  

### Primero tenemos que maximización respecto al control $c(t)$:  
La derivada parcial del Hamiltoniano respecto a $c(t)$ debe ser igual a cero:  
$  
\frac{\partial H}{\partial c(t)} = \frac{e^{-\rho t}}{c(t)} - \lambda(t) = 0.  
$  
De aquí, obtenemos la relación entre el consumo $c(t)$ y el multiplicador $\lambda(t)$:  
$  
\lambda(t) = \frac{e^{-\rho t}}{c(t)}.  
$  

### Ahora la ecuación de movimiento del estado $k(t)$:  
La ecuación de movimiento del capital es:  
$  
\dot{k}(t) = Ak(t) - c(t).  
$  

### La cuación adjunta para $\lambda(t)$:  
La derivada del multiplicador $\lambda(t)$ está dada por:  
$  
\dot{\lambda}(t) = \rho \lambda(t) - \frac{\partial H}{\partial k(t)}.  
$  
Dado que:  
$  
\frac{\partial H}{\partial k(t)} = \lambda(t) A,  
$  
entonces:  
$  
\dot{\lambda}(t) = \lambda(t)(\rho - A).  
$  

###   
La condición de transversalidad asegura que el capital no diverge en el largo plazo:  
$  
\lim_{t \to \infty} \lambda(t) k(t) = 0.  
$  

---  

### Entonces la solución es la siguiente: 

### (a) Solución para $\lambda(t)$:  
La ecuación diferencial para $\lambda(t)$ es:  
$  
\dot{\lambda}(t) = \lambda(t)(\rho - A).  
$  
Resolviendo esta ecuación diferencial, obtenemos:  
$  
\lambda(t) = \lambda(0) e^{(\rho - A)t}.  
$  

### (b) Relación entre $c(t)$ y $\lambda(t)$:  
De la relación $\lambda(t) = \frac{e^{-\rho t}}{c(t)}$, podemos despejar $c(t)$:  
$  
c(t) = \frac{1}{\lambda(0)} e^{-At}.  
$  

### (c) Dinámica del capital $k(t)$:  
Sustituyendo $c(t)$ en la ecuación de movimiento del capital:  
$  
\dot{k}(t) = Ak(t) - \frac{1}{\lambda(0)} e^{-At}.  
$  
La solución general de esta ecuación diferencial es:  
$  
k(t) = C e^{At} + \frac{1}{\lambda(0)} t e^{-At},  
$  
donde $C$ es una constante de integración.  

---  

## Los resultados son: 

### 1. Consumo óptimo:  
El consumo óptimo decrece exponencialmente con el tiempo:  
$  
c(t) = \frac{1}{\lambda(0)} e^{-At}.  
$  

### 2. Dinámica del capital:  
La evolución del capital está dada por:  
$  
k(t) = C e^{At} + \frac{1}{\lambda(0)} t e^{-At}.  
$  

### 3. Condición de transversalidad:  
La condición de transversalidad:  
$  
\lim_{t \to \infty} \lambda(t) k(t) = 0,  
$  
asegura que el capital no diverge en el largo plazo, garantizando la viabilidad económica del sistema.  

---  

## Ahora podemos concluir:

Este problema de optimización dinámica ilustra cómo un agente representativo elige su consumo $c(t)$ y acumula capital $k(t)$ a lo largo del tiempo para maximizar su utilidad intertemporal. El método del Hamiltoniano nos permite derivar las condiciones necesarias para encontrar la solución óptima, y las ecuaciones resultantes describen la evolución del consumo y el capital en el tiempo.  

## Este proceso se puede simular  con el siguiente código en python tom

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parámetros
rho = 0.03  # Tasa de descuento

A = 0.05    # Productividad del capital
lambda_0 = 1  # Valor inicial del multiplicador
k0 = 10     # Capital inicial

# Consumo óptimo
def c_optimal(t):
    return 1 / lambda_0 * np.exp(-A * t)

# Dinámica del capital
def capital_dynamics(t, k):
    c_t = c_optimal(t)  # Consumo en función del tiempo
    return A * k - c_t

# Resolver la ecuación diferencial para k(t)
t_span = (0, 100)  # Tiempo (de 0 a 100)
t_eval = np.linspace(0, 100, 1000)  # Puntos para evaluar
sol = solve_ivp(capital_dynamics, t_span, [k0], t_eval=t_eval)

# Extraer resultados
t = sol.t
k = sol.y[0]
c = c_optimal(t)

# Graficar resultados
plt.figure(figsize=(10, 6))
plt.plot(t, k, label="Capital $k(t)$", color="blue")
plt.plot(t, c, label="Consumo $c(t)$", color="green", linestyle="--")
plt.axhline(0, color="black", linewidth=0.5, linestyle=":")
plt.title("Dinámica del Capital y Consumo Óptimo")
plt.xlabel("Tiempo $t$")
plt.ylabel("Valor")
plt.legend()
plt.grid()
plt.show()
```
